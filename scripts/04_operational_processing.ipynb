```python
# 1. IMPORTS

from datetime import timedelta
import pandas as pd
import os
```

```python
# 2. PATHS

BASE_PATH = r"C:\Users\jujuf\Documents\projeto_cultura_inglesa"
OPERATIONAL_PATH = os.path.join(BASE_PATH, "data", "operational")
PROCESSED_PATH = os.path.join(BASE_PATH, "data", "processed")
os.makedirs(PROCESSED_PATH, exist_ok=True)
```

```python
# 3. LOAD DATAFRAME

cord_interventions = pd.read_excel(
    os.path.join(OPERATIONAL_PATH, "cord_interventions_forms.xlsx"),
    header=2
)
```

```python
display(cord_interventions)
```

```python
# 4. DATA TYPE TREATMENT

cord_interventions['Intervention_Date'] = pd.to_datetime(cord_interventions['Intervention_Date'])
```

```python
# 5. FILTER LAST 30 DAYS

today = pd.Timestamp('2026-02-12')
recent_interventions = cord_interventions[
    cord_interventions['Intervention_Date'] >= today - timedelta(days=30)
][['Student_ID', 'Intervention_Type', 'Intervention_Description', 'Case_Status', 'Intervention_Date']]
```

```python
# SAVE PROCESSED DATAFRAME

recent_interventions.to_csv(os.path.join(PROCESSED_PATH, "recent_interventions.csv"),index=False, encoding="utf-8-sig")
```
